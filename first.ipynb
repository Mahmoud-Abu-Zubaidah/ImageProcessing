```python
img = cv2.imread("./data/lion.jpg")
```

```python
img.shape
```

```python
img[:, :].shape
```

```python
plt.imshow(img[:,:,::-1])
plt.show()
#cv2 bgr in plt rgb
```

```python
# from rgb to hsv
imgHSV = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
imgHSV.shape
```

```python
plt.imshow(imgHSV[:,:,::-1])
plt.show()
```

```python
hue, sat, val = cv2.split(imgHSV)
```

```python
val.shape
```

```python
plt.imshow(hue)
plt.show()
```

```python
plt.imshow(sat)
plt.show()
```

```python
plt.imshow(val, cmap="gray")
plt.show()
```

```python
val.max(),val.min()
```

```python
val
```

```python
val = 255 - val
```

```python
val
```

```python
plt.imshow(val, cmap="gray")
plt.show()
```

```python
imgHSV = cv2.merge((val,val,val))
```

```python
plt.imshow(imgHSV[:,:,::-1])
plt.show()
```

```python
imgHSV = imgHSV[:,::-1,:]
```

```python
plt.imshow(imgHSV[:,:,::-1])
plt.show()
```

```python
imgNew = np.zeros((407, 640*2, 3), dtype=np.uint8)
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
imgNew[:,:640,:] = img
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
imgNew[:,640:,:] = imgHSV
```

```python
plt.imshow(imgNew[:,:,::-1])
plt.show()
```

```python
cv2.imwrite("result.png", imgNew)
```

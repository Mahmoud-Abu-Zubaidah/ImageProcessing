```python
gray_copy = cv2.cvtColor(screw,cv2.COLOR_BGR2GRAY)
```

```python
ret , binary_image = cv2.threshold(gray_copy,100,255,cv2.THRESH_OTSU)
```

```python
plt.imshow(binary_image,cmap='gray')
plt.show()
```

```python
contours , hist = cv2.findContours(binary_image, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
copy = np.copy(binary_image)

img = cv2.drawContours(screw, contours, 2, (0,0,255), 3)
```

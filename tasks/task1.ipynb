Q1: Image Translation -> shift of an image from one location to another

```python
#start code
image = np.zeros((400,500,3),dtype=np.uint8)
background = np.copy(bean[:200,:100])
background = cv2.resize(background[:,::-1],(500,400),interpolation=cv2.INTER_LINEAR)
image = np.copy(background)
image[400-373:,500-360:,:] = np.copy(bean)
plt.imshow(image[:,:,::-1])
#end code
```

Q2: Image Reflection -> flip the image horizontally & vertically

```python
#start code
vert = np.copy(bean[::-1,:])
vert = cv2.cvtColor(vert,cv2.COLOR_BGR2HSV)
vert[:,:,0] = vert[:,:,0] + 10
vert = cv2.cvtColor(vert,cv2.COLOR_HSV2BGR)
#------------------------------------------
#concat result in one image
#------------------------------------------
full_vert = np.zeros((373*2, 360, 3),dtype=np.uint8)
full_vert[:373,:,:] = bean[:,:,::-1]
full_vert[373:,:,:] = vert[:,:,::-1]

# horizantal
horiz = np.copy(bean[:,::-1,:])

full_horiz = np.zeros((373, 360*2, 3),dtype=np.uint8)
full_horiz[:,:360,:] = bean[:,:,::-1]
full_horiz[:,360:,:] = horiz[:,:,::-1]

plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(full_horiz)
plt.axis('off')

plt.subplot(122)
plt.imshow(full_vert)
plt.axis('off')

#end code
```
